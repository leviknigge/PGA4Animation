```python
def maxsmins(maxs, mins):
    i = 0
    j = 0
    xs = []
    ys = []
    
    while i < len(maxs) and j < len(mins):
        if maxs[i] < mins[j]:
            xs.append(maxs[i])
            ys.append(1)
            
            i = i+1
        else:
            xs.append(mins[j])
            ys.append(-1)
            
            j = j+1
    
    
    return xs, ys


xs, ys = maxsmins(maxs, mins)
needs = np.arange(462)
```
